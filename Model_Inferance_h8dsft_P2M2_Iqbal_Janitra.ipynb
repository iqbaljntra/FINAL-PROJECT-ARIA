```python
import pandas as pd
import numpy as np
import joblib
```

```python
with open('all_process', 'rb') as file_1:
  all_process= joblib.load(file_1)
```

```python
random = np.random.default_rng(seed=2895)
sample_type = random.choice(['lab 1','lab 2'], size=30)
v1 = random.integers(280, 600, size=30)
v2 = random.integers(230, 490, size=30)
v3 = random.integers(420, 670, size=30)
v4 = random.integers(300, 450, size=30)
v5 = random.integers(400, 700, size=30)
v6 = random.integers(200, 350, size=30)
v7 = random.integers(450, 900, size=30)
v8 = random.integers(3800, 4900, size=30)
data_inf = pd.DataFrame({'sample_type' : sample_type, 'v1' : v1, 'v2' : v2, 'v3':v3,
                        'v4' : v4, 'v5' : v5, 'v6' : v6, 'v7' : v7, 'v8' : v8})
data_inf
```

# MAKE A FINAL PREDICTION

```python
final_predict = all_process.predict(data_inf)
final_predict
```

```python
predict = pd.DataFrame(final_predict, columns=["Target Prediction"])
predict
```

```python
concatenated_df = pd.concat([predict, data_inf], axis=1)
concatenated_df
```
